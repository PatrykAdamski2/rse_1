```python
import pint
import ballistics
import matplotlib.pyplot as plt
import numpy as np

ureg = pint.UnitRegistry()
```

## Trajectories for several launch angles

```python
v0 = ureg.Quantity(80, 'm/s')
angles = [15, 30, 45, 60, 75]

fig, ax = plt.subplots(figsize=(10, 5))

for deg in angles:
    angle = ureg.Quantity(deg, 'degree')
    t, x, y = ballistics.trajectory(angle, v0)
    r = ballistics.max_range(angle, v0).to('m').magnitude
    h = ballistics.max_height(angle, v0).to('m').magnitude
    ax.plot(
        x.to('m').magnitude,
        y.to('m').magnitude,
        label=f'{deg}°  (R={r:.0f} m, H={h:.0f} m)'
    )

ax.set_xlabel('Horizontal distance (m)')
ax.set_ylabel('Height (m)')
ax.set_title(f'Projectile trajectories  |  v₀ = {v0}')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

## Range vs. angle — finding the optimum

```python
v0 = ureg.Quantity(100, 'm/s')
angle_deg = np.linspace(1, 89, 200)
ranges = [
    ballistics.max_range(ureg.Quantity(a, 'degree'), v0).to('m').magnitude
    for a in angle_deg
]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(angle_deg, ranges)
ax.axvline(45, color='red', linestyle='--', label='45° optimum')
ax.set_xlabel('Launch angle (°)')
ax.set_ylabel('Range (m)')
ax.set_title('Range as a function of launch angle')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

## Unit conversion example

Pint handles any combination of units transparently.

```python
angle = ureg.Quantity(45, 'degree')
v0    = ureg.Quantity(200, 'km/h')

r = ballistics.max_range(angle, v0)
h = ballistics.max_height(angle, v0)

print(f'v₀ = {v0} = {v0.to("m/s"):.2f}')
print(f'Range  = {r.to("m"):.1f}  =  {r.to("km"):.3f}')
print(f'Height = {h.to("m"):.1f}')
```
